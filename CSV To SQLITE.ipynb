```python
# Dependencies
import csv, sqlite3
con = sqlite3.connect("Sales.db")
cur = con.cursor()
cur.execute("CREATE TABLE Sales_table (Rank, Name, Platform, Year, Genre,Publisher, NA_Sales, EU_Sales,JP_Sales,Other_Sales,Global_Sales);") # use your column names here
```

```python

with open('data.csv','r') as fin: # `with` statement available in 2.5+
    # csv.DictReader uses first line in file for column headings by default
    dr = csv.DictReader(fin) # comma is default delimiter
    to_db = [(i['Rank'], i['Name'], i['Platform'],i['Year'],i['Genre'], i['Publisher'],i['NA_Sales'],i['EU_Sales'], i['JP_Sales'],i['Other_Sales'],i['Global_Sales']) for i in dr]
to_db
```

```python
cur.executemany("INSERT INTO Sales_table (Rank, Name, Platform, Year, Genre, Publisher, NA_Sales, EU_Sales, JP_Sales, Other_Sales, Global_Sales) VALUES (?, ?,?,?,?,?,?,?,?,?,?);", to_db)
con.commit()
con.close()
```
